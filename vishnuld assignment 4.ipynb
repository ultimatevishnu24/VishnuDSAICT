```python

#importing pandas library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import MinMaxScaler
 
#loading data
file = pd.read_csv(r"C:\Users\hp\Downloads\titanic.csv")
#display the file
print(file)
print('\n')
file.head()
print('\n')
#creating the dataframe
df = pd.DataFrame(file)
#display the dataframe
df
print('\n')
#setting passengerid as index
df.set_index('PassengerId')
print('\n')
#checking the information in the data frame
df.info
# using isnull() function to identify the null values
df.isnull()
print('\n')
# to interpolate the missing values, fill the voids
df.interpolate(method ='linear', limit_direction ='forward')
print('\n')

#to identify the ouliers, using boxplot 
#boxplot for column 'age'
sns.boxplot(df['Age'])
plt.show()
#identiify the ouliers in boxplot for coulmn 'Age'
print(np.where(df['Age']>60))
print('\n')

#boxplot for column 'Fare'
sns.boxplot(df['Fare'])
plt.show()
#identiify the ouliers in boxplot for coulmn 'Age'
print(np.where(df['Fare']>100))
print('\n')


scaler = MinMaxScaler()
# transform data
scaled = scaler.fit_transform(df[["Survived"]])
print(scaled)
```
