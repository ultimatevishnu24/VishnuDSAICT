```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#to read the excel file
file = pd.read_excel(r"C:\Users\Dell\Desktop\iris.xls")
file.head()

#to display the data and checking
print(file,'\n')
print(file["Classification"].value_counts(),'\n')

#to calculate the mean of each column in the data set
m1=file['SL'].mean()
m2=file['SW'].mean()
m3=file['PL'].mean()
m4=file['PW'].mean()

#to print the mean of each column in dataset
print("mean of SL is",m1,'\n', "mean of SW is",m2,'\n',"mean of PL is",m3,'\n',"mean of PW is",m4)

#to check for null values
v1=file['SL'].isnull().values.any()
v2=file['SW'].isnull().values.any()
v3=file['PL'].isnull().values.any()
v4=file['PW'].isnull().values.any()

#to print whether there are null values
print(v1,'\n',v2,'\n',v3,'\n',v4)

#creating data for plotting
df1=file[0:50]
df1m1=df1['SL'].mean()
df1m2=df1['SW'].mean()
df1m3=df1['PL'].mean()
df1m4=df1['PW'].mean()

df2=file[50:100]
df2m1=df2['SL'].mean()
df2m2=df2['SW'].mean()
df2m3=df2['PL'].mean()
df2m4=df2['PW'].mean()


df3=file[101:150]
df3m1=df3['SL'].mean()
df3m2=df3['SW'].mean()
df3m3=df3['PL'].mean()
df3m4=df3['PW'].mean()

plotdata={'SL': [df1m1, df2m1, df3m1],
        'SW': [df1m2, df2m2, df3m2],
        'PL':[df1m3,df2m3,df3m3],
        'PW':[df1m4,df2m4,df3m4],
        'Species':["iris_setosa","Iris_versicolor","Iris_virginica"]}

df=pd.DataFrame(plotdata)
print('\n')
print(df,'\n')

# scatter plot
df.plot(kind = 'scatter',
        x = 'Species',
        y = 'PW',
        color = 'red')
  
# set the title
plt.title('ScatterPlot')
  
# show the plot
plt.show()

# bar plot
df.plot(kind = 'bar',
        x = 'Species',
        y = 'SW',
        color = 'green')
  
# set the title
plt.title('BarPlot')
  
# show the plot
plt.show()

#Get current axis
ax = plt.gca()

# line plot 
df.plot(kind = 'line',
        x = 'Species',
        y = 'SL',
        color = 'green',ax = ax)
  
# line plot 
df.plot(kind = 'line',
         x = 'Species',
        y = 'SW',
        color = 'blue',ax = ax)
  
# line plot 
df.plot(kind = 'line',
        x = 'Species',
        y = 'PL',
        color = 'black',ax = ax)

df.plot(kind = 'line',
        x = 'Species',
        y = 'PW',
        color = 'black',ax = ax)
  
# set the title
plt.title('LinePlots')
  
# show the plot
plt.show()
```
